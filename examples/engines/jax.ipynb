What is horqrux?

horqrux is the JAX-equivalent of pyqtorch for qadence, for more info, check out https://pasqal-io.github.io/horqrux/latest/.

But what is JAX?

JAX (https://jax.readthedocs.io/en/latest/) is a high-performance array computing library combining automatic differentation and XLA (Accelerated Linear Algebra), an open-source compiler for machine learning. 

It's heavily functional programming- and NumPy-inspired. JAX arrays can often be used as drop-in replacements of NumPy arrays. However, unlike NumPy arrays, JAX arrays are always immutable.


Why use JAX?

0. Functional Programming paradigm
1. Jit and XLA
2. Auto-vectorization and parallelization using vmap/pmap
3. Pytrees and tree_map: Using arbitary Python containers for computation
4. Control over compiler-level operations
4. Efficient Differentiation of any pure Python function

What are the main differences to torch?

0. Functional, instead of object-ortiented (stateful) interface
1. No 'impure' operations, i.e., in-place updates (mutation)
2. Pseudorandom number generators (PRNGs) via explicitly passing and iterating PRNG state
3. No functions with argument-value dependent shapes
4. Compiler-level control-flow primitives

How do i use it?

`pip install qadence[horqrux]`

For GPU support

`pip install --upgrade "jax[cuda11_pip]" -f https://storage.googleapis.com/jax-releases/jax_cuda_releases.html`

Can I use it on the cluster?

Yes, however you need to `pip install qadence[horqrux]` and JAX manually.


Should use it?

At the moment, horqrux is only available via the low-level backend API, so it's recommended for qadence-experienced devs.
